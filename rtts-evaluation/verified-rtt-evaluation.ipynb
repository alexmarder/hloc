# RTT Evaluation of Verified IP Addresses

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
from numpy import genfromtxt
TUMDarkerBlue="#004b87"
TUMLighterBlue="#92c1e9"
TUMBlue="#005eb8"
print("Numpy version: {}".format(np.__version__))
import sys, subprocess
print("Python version: {}".format(sys.version))
maxX =500
maxY = 100
```

```python
PATH="../rtts-evaluation/" # modify PATH to point to the rtts-evaluation dataset
```

## IPv6 only

```python
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv6-woip-wodist.rtts')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv6-wip-wodist.rtts'), v6_ver_rtt_data)

v6_ver_wod_sorted_rtt_data = np.sort([rtt * 100 for rtt in v6_ver_rtt_data])
v6_ver_wod_y_values_rtt_data = np.arange(len(v6_ver_wod_sorted_rtt_data))/ \
    float(len(v6_ver_wod_sorted_rtt_data))*100
    
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv6-woip.dist')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv6-wip.dist'), v6_ver_rtt_data)
v6_ver_d_sorted_rtt_data = np.sort([dist/2 for dist in v6_ver_rtt_data])
v6_ver_d_y_values_rtt_data = np.arange(len(v6_ver_d_sorted_rtt_data))/ \
    float(len(v6_ver_d_sorted_rtt_data))*100
       
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv6-woip-summed.dist')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv6-wip-summed.dist'), v6_ver_rtt_data)
v6_ver_dn_sorted_rtt_data = np.sort(v6_ver_rtt_data)
v6_ver_dn_y_values_rtt_data = np.arange(len(v6_ver_dn_sorted_rtt_data))/ \
    float(len(v6_ver_dn_sorted_rtt_data))*100

plt.plot(v6_ver_wod_sorted_rtt_data, v6_ver_wod_y_values_rtt_data, color='blue')
plt.plot(v6_ver_d_sorted_rtt_data, v6_ver_d_y_values_rtt_data, color='red')
plt.plot(v6_ver_dn_sorted_rtt_data, v6_ver_dn_y_values_rtt_data, color='green')

plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%d %%'))
plt.axis([0, 3000, 0, maxY])
plt.xlabel('Round Trip Time')
plt.show()
```

## IPv4 only

```python
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv4-woip-wodist.rtts')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv4-wip-wodist.rtts'), v6_ver_rtt_data)

v4_ver_wod_sorted_rtt_data = np.sort([rtt * 100 for rtt in v6_ver_rtt_data])
v4_ver_wod_y_values_rtt_data = np.arange(len(v4_ver_wod_sorted_rtt_data))/ \
    float(len(v4_ver_wod_sorted_rtt_data))*100
    
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv4-woip.dist')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv4-wip.dist'), v6_ver_rtt_data)

v4_ver_d_sorted_rtt_data = np.sort([dist/2 for dist in v6_ver_rtt_data])
v4_ver_d_y_values_rtt_data = np.arange(len(v4_ver_d_sorted_rtt_data))/ \
    float(len(v4_ver_d_sorted_rtt_data))*100
       
v6_ver_rtt_data = genfromtxt(PATH+'verified/ipv4-woip-summed.dist')
v6_ver_rtt_data = np.append(genfromtxt(PATH+'verified/ipv4-wip-summed.dist'), v6_ver_rtt_data)

v4_ver_dn_sorted_rtt_data = np.sort(v6_ver_rtt_data)
v4_ver_dn_y_values_rtt_data = np.arange(len(v4_ver_dn_sorted_rtt_data))/ \
    float(len(v4_ver_dn_sorted_rtt_data))*100

plt.plot(v4_ver_wod_sorted_rtt_data, v4_ver_wod_y_values_rtt_data, color='blue')
plt.plot(v4_ver_d_sorted_rtt_data, v4_ver_d_y_values_rtt_data, color='red')
plt.plot(v4_ver_dn_sorted_rtt_data, v4_ver_dn_y_values_rtt_data, color='green')

plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%d %%'))
plt.axis([0, 3000, 0, maxY])
plt.xlabel('Round Trip Time')
plt.show()
```

# IPv4 + IPv6 (used for paper)

```python
v6_ver_rtt_data = np.append(v4_ver_wod_sorted_rtt_data, v6_ver_wod_sorted_rtt_data)
v4_ver_wod_sorted_rtt_data = np.sort(v6_ver_rtt_data)
v4_ver_wod_y_values_rtt_data = np.arange(len(v4_ver_wod_sorted_rtt_data))/ \
    float(len(v4_ver_wod_sorted_rtt_data))*100
    
v6_ver_rtt_data = np.append(v4_ver_d_sorted_rtt_data, v6_ver_d_sorted_rtt_data)
v4_ver_d_sorted_rtt_data = np.sort(v6_ver_rtt_data)
v4_ver_d_y_values_rtt_data = np.arange(len(v4_ver_d_sorted_rtt_data))/ \
    float(len(v4_ver_d_sorted_rtt_data))*100
    
v6_ver_rtt_data = np.append(v4_ver_dn_sorted_rtt_data, v6_ver_dn_sorted_rtt_data)
v4_ver_dn_sorted_rtt_data = np.sort(v6_ver_rtt_data)
v4_ver_dn_y_values_rtt_data = np.arange(len(v4_ver_dn_sorted_rtt_data))/ \
    float(len(v4_ver_dn_sorted_rtt_data))*100
    
plt.plot(v4_ver_d_sorted_rtt_data, v4_ver_d_y_values_rtt_data, color=TUMLighterBlue, label="(1) x = Distance Probe - Hint ")
plt.plot(v4_ver_wod_sorted_rtt_data, v4_ver_wod_y_values_rtt_data, linestyle="dashed", color=TUMBlue, label="(2) a = Latency Buffer Used")
plt.plot(v4_ver_dn_sorted_rtt_data, v4_ver_dn_y_values_rtt_data, color=TUMBlue, label="(3) Maximum Error (2x+a)")

plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%d %%'))
plt.axis([0, 3000, 0, maxY])
plt.legend(loc="lower right")
#plt.title('CDF for minimum RTT from any server')
plt.xlabel("km")
plt.ylabel("% Domains [x < X]")
plt.xlim(0,1500)

plt.savefig('verified_rtts_eval.pdf', format='pdf', dpi=2000)
plt.show()
```
